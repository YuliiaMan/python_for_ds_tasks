### Профіль колонок набору даних:

- **datetime**: дата та час зібрані погодинно

  1. Ясно, Мало хмарно, частково хмарно

- **atemp**: як відчувається температура в градусах Цельсія

- **windspeed**: швидкість вітру
- **casual**: кількість неформальних користувачів
- **registered**: кількість зареєстрованих користувачів
- **count**: загальна кількість велосипедів, взятих в оренду, включаючи неформальних та зареєстрованих користувачів

```python
spring_summer_season = df[df['season'].isin([1, 2])]
spring_summer_season
```

На проміжку користувачів від 200 до 400, кількість випадкових користувачів продовжує свою активність на проміжку від 50 до 300.
Неформальні користувачі важливі, оскільки вони забезпечують додатковий прибуток у пікові періоди, такі як вихідні, свята або теплі сезони. Таким чином неформальні користувачі підсилюють кількість звичайних користувачів, збільшуючи прибуток компанії.

 Тепер давайте візуалізуємо кількість зареєстрованих та казуальних користувачів поденно. Для цього виконайте код нижче і після цього створіть нову змінну `df_viz` яка містить лише колонки `'date', 'registered', 'casual'`, згрупуйте дані за датою і відобразіть на **одному** графіку поденну динаміку двох показників з допомогою `pandas.plot`.

```python
df['date'] = df.datetime.dt.date
```

```python
df_viz = df[['date', 'registered', 'casual']].groupby('date').sum().reset_index()
df_viz
```

```python
plt.figure(figsize=(8, 6))
plt.plot(df_viz['date'], df_viz['registered'], label='registered', color='blue')
plt.plot(df_viz['date'], df_viz['casual'], label='casual', color='orange')
plt.xlabel('date')
plt.legend()
plt.show();
```
